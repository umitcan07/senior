```python
import librosa
import soundfile as sf

# Load the MP3 file and convert to mono
# librosa.load automatically converts to mono by default
audio_path = "audio/pi.mp3"
audio, sr = librosa.load(audio_path, sr=None, mono=True)

# Save as mono WAV file
output_path = "audio/pi_mono.wav"
sf.write(output_path, audio, sr)

print(f"Converted {audio_path} to mono WAV: {output_path}")
print(f"Sample rate: {sr} Hz")
print(f"Duration: {len(audio) / sr:.2f} seconds")
print(f"Channels: 1 (mono)")
```

```python
import librosa
import soundfile as sf

# Load the mono WAV file
audio_path = "audio/pi_mono.wav"
audio, sr = librosa.load(audio_path, sr=None, mono=True)

# Trim from 1.10 to 2.00 seconds
start_time = 70  # seconds
end_time = 120    # seconds

start_sample = int(start_time * sr)
end_sample = int(end_time * sr)

trimmed_audio = audio[start_sample:end_sample]

# Save the trimmed audio
output_path = "audio/pi_mono_trimmed.wav"
sf.write(output_path, trimmed_audio, sr)

print(f"Trimmed audio from {start_time}s to {end_time}s")
print(f"Output saved to: {output_path}")
print(f"Duration: {len(trimmed_audio) / sr:.2f} seconds")
```
